```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read csv
import pandas as pd 
wine_df = pd.read_csv("winedata.csv")
wine_df.head()
```

```python
#split data into our features and target arrays 
features = wine_df.drop(columns='type', axis=1).values
target = wine_df['type'].values

#split data into a training and test dataset
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=42)
```

```python
#fit using StandardScaler
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)

#scale data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Define the model: 
nn = tf.keras.models.Sequential()

#First Hidden Layer
nn.add(tf.keras.layers.Dense(units=16, activation="relu", input_dim=13))

#Second Hidden Layer
nn.add(tf.keras.layers.Dense(units=16, activation="relu"))

#Output Layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

#Checking the structure
nn.summary()
```

```python
#Compile the Model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
#Train the Model 
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose='auto')
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
